# Class Task
### Read random wikipedia pages, tokenize, lemmatize and create a word cloud.
STEPS:
1. Read Data
2. Make string lower
3. Clean Data (Select only words)
4. Tokenize
5. Remove Stop words
6. Lemmatize
7. Generate Frequency distribution plot (code provided)

```python
!pip install wikipedia
import wikipedia
import wordcloud
import matplotlib.pyplot as plt

# title = wikipedia.random()
title='Kargil'
wikipedia.search(title)
page=wikipedia.page(title)
content = page.content

print(title)
```

```python
content
```

# STRING LOWER CASE

```python
content=content.lower()
content
```

**-----------------------------------------------------------------------------------------------------------------------------------------------------------
-----**

# TOKENIZATION

```python
nltk.download('punkt')
tokenizer=nltk.tokenize.word_tokenize
tokens= tokenizer(content)
print(tokens)
```

**-----------------------------------------------------------------------------------------------------------------------------------------------------------
-----**

# DATA CLEANING

```python
import re
abc = re.findall(r'[a-z]+',content)
content = abc
```

**-----------------------------------------------------------------------------------------------------------------------------------------------------------
-----**

# REMOVING STOPWORDS

```python
nltk.download('stopwords')
en=set(nltk.corpus.stopwords.words('english'))
```

```python
final_content=[]
for i in content:
  if i not in en:
    final_content.append(i) 
```

# PERFORMING LEMITIZATION

```python
nltk.download('wordnet')
import nltk
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer
 
lemmatizer = WordNetLemmatizer()

words = []
 
for w in final_content:
    words.append(lemmatizer.lemmatize(w))
```

# FREQUENCY DISTRIBUTION

```python
from nltk import FreqDist

freq=FreqDist(words)
print(freq.most_common())

freq.plot(20)
```

```python
wc=wordcloud.WordCloud(width = 800, height = 800,
                background_color ='white', min_font_size=10).generate_from_frequencies(freq)
plt.figure()

plt.imshow(wc)
```
